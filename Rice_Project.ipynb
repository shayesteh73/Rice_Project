```python
# extracting a column from a dataset

rice_df = df.drop(['Class'], axis=1)
rice_df[0:10]
```

```python
# standardization

df_scaled = pd.DataFrame({'Area':[], 'Perimeter':[], 'Major_Axis_Length':[],'Minor_Axis_Length':[],'Eccentricity':[],'Convex_Area':[],'Extent':[]})

for i in ['Area','Perimeter','Major_Axis_Length','Minor_Axis_Length','Eccentricity','Convex_Area','Extent']:
    array = []
    mean = rice_df[i].mean()
    std = rice_df[i].std()
    for x in rice_df[i]:
      array.append((x - mean)/std)
    df_scaled[i] = array
df_scaled.head()
```

```python
# converting a new data to a numerical array

np_test_scaled = test_data_scaled.to_numpy()
np_test_scaled
```

```python
# converting raw data to numerical arrays

np_df_scaled = df_scaled.to_numpy()
np_df_scaled
```

```python
# Calculating euclidean distance

euclidean_dist = np.zeros(shape=(len(np_df_scaled)))
euclidean_dist
```

```python
for i in range(len(np_df_scaled)):
  euclidean_dist[i ] = np.sqrt(np.sum([(a-b)*(a-b) for a,b in zip(np_test_scaled,np_df_scaled[i])]))
print(euclidean_dist)
```

```python
# Adding spaces as a column in a dataframe

df_scaled['euclidean_dist'] = euclidean_dist
df_scaled.head()
```

```python
# Adding the labels we had set aside to the standardized data

df_scaled['Class'] = df['Class']
df_scaled.head()
```

```python
# Sorting the data based on the Euclidean distance from the new data

sorted_df = df_scaled.sort_values(by=['euclidean_dist'])
sorted_df
```
